```python
import requests
import os
import json

# URL to fetch the metadata
get_url = "https://api-production.data.gov.sg/v2/public/api/collections/189/metadata"

# Directory to save the files
save_dir = "downloaded_files"

# Create the directory if it doesn't exist
if not os.path.exists(save_dir):
    os.makedirs(save_dir)

# Fetch the metadata
response = requests.get(get_url)
data = response.json()
id_list = data['data']['collectionMetadata']['childDatasets']

# Iterate through each dataset ID and save the data
for i in id_list:
    base_url = "https://data.gov.sg/api/action/datastore_search"
    url = base_url + f'?resource_id={i}'
    
    # Fetch the data from the URL
    dataset_response = requests.get(url)
    dataset_data = dataset_response.json()
    
    # Save the data to a file
    file_path = os.path.join(save_dir, f'dataset_{i}.json')
    with open(file_path, 'w') as file:
        json.dump(dataset_data, file)
    
    print(f'Saved data to {file_path}')
```

```python
import os
import pandas as pd

path = "C:/Users/Hari Haran/OneDrive/Desktop/GUVI/Project_1 _youtube_harvesting/Project_3_Singapore_Resale_Falt_Process_Prediction/data"
files = os.listdir(path)
file_names = []

for i in files:
    file_path = os.path.join(path, i)
    if os.path.isfile(file_path) and file_path.endswith('.csv'):  # Check if the file is a CSV
        data = pd.read_csv(file_path)
        # Use a valid Python variable name by replacing invalid characters
        variable_name = i.replace('.', '_').replace('-', '_') + '_df'
        globals()[variable_name] = data
        file_names.append(i)

# Print the file names
for file_name in file_names:
    print(file_name)
```
